# Multiple Regression

```python
import numpy as np
import pandas as pd
import sklearn
import sys

print("Numpy version :", np.__version__)
print("Pandas version :", pd.__version__)
print("System version :", sys.version)
print("Sklearn version :", sklearn.__version__)
```

```python
from sklearn.datasets import load_boston
boston_data = load_boston()
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
df.head()
```

```python
df.shape
```

```python
# X or feature will be the whole column in the dataset other than target MEDV
X = df
y = y = boston_data.target
```

# Using Statmodels

```python
import statsmodels.api as sm
X_constant = sm.add_constant(X)
pd.DataFrame(X_constant)
```

```python
model = sm.OLS(y, X_constant)
lr = model.fit()
lr.summary()
# look at R-squared for model accuracy, P value for feature significant and coef value
```

# Using Statmodels Formula API

```python
import statsmodels.formula.api as smf

form_lr = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data=df)
mlr = form_lr.fit()

mlr.summary()

# using formula API we can remove feature with p value higher than 0.05 (remove using formula equation above)
```

```python
form_lr = smf.ols(formula = 'y ~ CRIM + ZN + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT', data=df)
mlr = form_lr.fit()

mlr.summary()
```

# Use $R^2$ to Identify Key Features

```python
from sklearn.metrics import r2_score
linear_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', 
              data=df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
# without LSTAT
linear_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B', 
              data=df)
lr_without_LSTAT = linear_reg.fit()
r2_score(y, lr_without_LSTAT.predict(df))
```

```python
# without AGE
linear_reg = smf.ols(formula = 'y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT', 
              data=df)
lr_without_AGE = linear_reg.fit()
r2_score(y, lr_without_AGE.predict(df))
```
